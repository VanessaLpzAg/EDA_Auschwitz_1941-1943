## 1. CONTEXTO.

La Segunda Guerra Mundial fue un conflicto bélico global que tuvo lugar entre el 1 de Septiembre de 1939 y el 2 de Septiembre de 1945, involucrando a la mayoría de las naciones del mundo. Fue la guerra más extensa y devastadora de la historia, con combates en Europa, África, Asia y el Pacífico.

La guerra fue causada por una combinación de factores, como el descontento con el Tratado de Versalles (que puso fin a la Primera Guerra Mundial), la expansión imperialista de Alemania, Italia y Japón, y la política de apaciguamiento de las democracias europeas frente a las agresiones de las potencias del Eje (Alemania, Italia y Japón). La invasión de Polonia por parte de Alemania, junto con la violación de los acuerdos internacionales, marcó el inicio formal del conflicto.

La guerra dejó un saldo de decenas de millones de muertos y alteró por completo el orden político y económico mundial.

Dentro de este contexto nos encontramos con la creación de los campos de concentración, que fueron construidos por Alemania desde 1933. Se encarcelaron en ellos a los oponentes políticos del régimen nazi y a la gente considerada como “elementos indeseables”.

Los denominados "elementos indeseables" eran las personas que no se adecuaban a la ideología nazi, con vistas a crear una sociedad “racialmente pura”, los nazis alemanes planearon el exterminio de los judíos, los eslavos, romaníes (gitanos), homosexuales y otros.

El campo de concentración nazi de *Auschwitz* se ha convertido, para el mundo, en el símbolo del Holocausto, genocidio y terror. Fue construido por los alemanes a mediados de 1940, en las afueras de *Oświęcim*, ciudad polaca que había sido anexionada al III Reich.

El motivo directo de la apertura del campo fue el creciente número de polacos arrestados por la policía alemana y su correspondiente hacinamiento en las cárceles. Será a partir de 1942, cuando empezó a convertirse progresivamente en el mayor centro de  exterminio masivo, convirtiéndose en un complejo de tres campos principales: *Auschwitz I* (campo principal), *Auschwitz II-Birkenau* (campo de exterminio) y *Auschwitz III-Monowitz* (campo de trabajo).

*Auschwitz II-Birkenau* fue el lugar donde ocurrieron las mayores atrocidades, incluyendo el genocidio de aproximadamente 1.1 millones de personas. Las víctimas fueron asesinadas en cámaras de gas, y los cuerpos incinerados en crematorios. 

El campo fue liberado por el Ejército Soviético el 27 de enero de 1945. Hoy en día, *Auschwitz* es un memorial y museo en honor a las víctimas del Holocausto.

## 3. OBTENCIÓN DE LOS DATOS.

### 3.1. Fuente de los Datos.

Los datos obtenidos para realizar el Análisis Exploratorio de Datos (EDA) se obtienen en la página web del [*United States Holocaust Memorial Museum*](https://www.ushmm.org/es), a través de la Base de Datos que recogen diversa documentación.

Esta Base de Datos contiene los llamados "Libros de la muerte" (*Sterbebücher*) (ID: 49478), parcialmente conservados, que recogen los certificados de defunción de los prisioneros del campo de concentración de *Auschwitz*. Son 46 volúmenes de defunciones de prisioneros que fueron registrados en el campo y murieron entre el 29 de julio de 1941 y el 31 de diciembre de 1943. La información proporcionada es apellido, nombre(s), fecha de nacimiento, fecha de defunción, lugar de nacimiento, lugar de residencia y religión.

Al utilizar la base de datos, hay que tener presente que los certificados de defunción sólo se refieren a los prisioneros registrados que murieron en el período mencionado en el párrafo anterior. La inmensa mayoría de las víctimas, perecieron en las cámaras de gas inmediatamente después de su llegada, sin que se las incluyera en los registros del campo y sin que sus muertes se anotaran en los documentos alemanes.

#### 7.2.7. Religiones.

```python
'''Crear un Dataframe de religiones presentes.'''

df_religiones = pd.DataFrame(df['Religion'].value_counts()).reset_index()
df_religiones
```

```python
'''Definir los grupos de religiones a combinar para su mejor análisis.'''

# Crear diccionario, para agrupar las religiones y ponerles nuevo nombre.
religiones_agrupadas = {
    'Catholic': ['Catholic', 'Greek Catholic'],
    'Jew': ['Jew'],
    'Orthodox' : ['Russian Orthodox', 'Eastern Orthodox', 'Greek Orthodox'],
    'Protestant' : ['Protestant'],
    'Atheism': ['Atheist', 'Agnostic'],    
    'Other Religions': ['Muslim', 'Buddhist', 'Jehovah\'s Witness', 'Hussite', 'Czech-Moravian'],
    'Unknown': ['Unknown', 'Unaffiliated', 'Believes in God', 'andere']
}

# Función para asignar un grupo a cada religión.
def asignar_religion_a_grupo(religion):
    for group, religions in religiones_agrupadas.items():
        if religion in religions:
            return group
    return religion  # En caso de que no se encuentre en ningún grupo, se deja tal cual.

# Aplicar la función a la columna 'Religión' para agrupar.
df_religiones['Religion'] = df_religiones['Religion'].apply(asignar_religion_a_grupo)

# Agrupar por 'Religion Group' y sumar los conteos.
df_religion = pd.DataFrame(df_religiones.groupby('Religion')['count'].sum().reset_index())

# Ordenar por conteo de mayor a menor.
df_religion= df_religion.sort_values(by='count', ascending=False).reset_index(drop=True)

# Mostrar el resultado.
df_religion
```

```python
'''Visualización de las Religiones presentes.'''

# Lista de colores.
c_7 = ['#F7E1A5', '#B8E2D6', '#A7C7E7', '#F7C6D0', '#D3B9E3', '#F8C2B7', '#A7E7C5']
#Crear grafico.
desfase = (0.1, 0.1, 0, 0, 0, 0, 0)
plt.pie(df_religion['count'], autopct="%0.1f %%", colors=c_7, explode=desfase);
plt.axis("equal");
plt.title('Religiones presentes entre los prisioneros');
plt.legend(df_religion['Religion'], shadow=False, fontsize="7", loc='lower right')
sns.set_style("whitegrid");
```

Este gráfico nos muestra la representación de las religiones presentes entre los prisioneros, llama la atención que el porcentaje de católicos y judíos en estos años se encuentre igualado, pero tiene su explicación histórica. En este momento, Alemania se encontraba consolidando sus territorios ocupados (Francia, Polonia, Bélgica, Luxemburgo, etc.) de mayoría católica, aunque también con presencia judía en menor medida, pero este grupo de población sufría la existencia de la operación 'Silencio Total', por lo que es entendible que los porcentajes se igualen. Además, Alemania se encontraba en pleno ataque a la Unión Soviética, el cual inició en 1942, observándose en ese porcentaje de ortodoxos en cuarta posición. 
Como dato adicional, se produjo el traslado mayoritario de judíos al campo con las deportaciones de 1944 desde Hungría, datos que nos quedan fuera de este *data set*.
